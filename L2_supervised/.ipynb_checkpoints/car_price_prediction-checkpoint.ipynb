# One hot encoding and label encoding

```python
X = df_cleaned.drop(columns="price",axis=1)
y=df_cleaned['price']
```

```python
df_cleaned.columns
```

```python
X_one_encode = pd.get_dummies(X,columns=['model','transmission','fuelType'],drop_first=True)
```

```python
X_one_encode
```

```python
X_one_encode = X_one_encode.astype(int)
X_one_encode.head()
```

```python
from sklearn.preprocessing import LabelEncoder

columns = ['model','transmission','fuelType']
Xlabel = X.copy()
label_encoder={}
for i in columns:
    le = LabelEncoder()
    Xlabel[i]=le.fit_transform(Xlabel[i])
    label_encoder[i] = le
```

```python
Xlabel
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
numeric_cols = ['year','mileage','tax','mpg']
scaler = StandardScaler()
X_one_encode[numeric_cols] = scaler.fit_transform(X_one_encode[numeric_cols])
X_one_encode
```

```python
Xlabel.columns
```

```python
xlabel_columns = ['model', 'year', 'transmission', 'mileage', 'fuelType', 'tax', 'mpg','engineSize']
Xlabel[xlabel_columns] = scaler.fit_transform(Xlabel[xlabel_columns])
```

```python
Xlabel
```

# Training , testing and model accuracy

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_one_encode, y, test_size=0.33, random_state=42)
```

```python
model = LinearRegression()
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
y_test
```

```python
r2 = r2_score(y_test,y_pred)
r2 #model accuracy
```

```python
#total rows in X_test
n=X_test.shape[0]
p=X_test.shape[1] #total cols in X_test

adjusted_r2 = 1 - ((1-r2)*(n-1))/(n-p-1)
adjusted_r2
```

# training testing for labeled data

```python
X_train, X_test, y_train, y_test = train_test_split(Xlabel, y, test_size=0.33, random_state=42)
```

```python
model2 = LinearRegression()
model2.fit(X_train,y_train)
```

```python
y_pred = model2.predict(X_test)
y_pred
```
